```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


df = pd.read_csv("https://raw.githubusercontent.com/evanzs/Curso_DataScience_Minsait/main/Datasets/populacao_brasileira.csv", sep = ";")
df_german = pd.read_csv("https://raw.githubusercontent.com/evanzs/Curso_DataScience_Minsait/main/Datasets/german_credit_data.csv", sep = ",")
```

```python
1. #**Gráfico de Linha:**
   #Crie um gráfico de linha mostrando o crescimento populacional ao longo dos anos (populacao_brasileira.csv) ou variação de idades ao longo dos registros (german_credit_data.csv).
df['year'] = df['year'].astype(int)
df['population'] = df['population'].astype(int)


plt.subplot(1, 2, 1)
plt.plot(df['population'],df['year'])
plt.show()


plt.subplot(1, 2, 2)
plt.plot(df_german.['Age'].count,df_german['Age'])
plt.show()


```

```python
#2. **Gráfico de Barras:**
 #  Represente a distribuição de frequência de variáveis qualitativas (categorias) (por exemplo, pessoas por sexo).
plt.subplot(1, 2, 1)

german = df_german['Sex'].value_counts()

plt.bar(german.index,german.values)
plt.show()

```

```python
#1. **Gráfico de Pizza:**
 #  Mostre a distribuição percentual de variáveis categóricas (por exemplo, distribuição dos propósitos).
df_german

Purpose_info = df_german['Purpose'].value_counts()

fig = plt.figure(figsize =(10, 7))
plt.pie(Purpose_info.values, labels = Purpose_info.index,autopct='%1.1f%%',textprops=dict(color='w'))
plt.show()
```

```python
#2. **Histograma:**
#   Crie um histograma para analisar a distribuição de uma variável (por exemplo, idade) em um conjunto de dados.
age_info = df_german['Age'].value_counts()

plt.figure(figsize=(16,8))

#bins = tamanho das barras
plt.hist(df_german['Age'],bins=20,color='skyblue',edgecolor='black')
plt.title("Distribuição de idades")
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.show()
```

```python
#3. **Gráfico de Dispersão:**
#   Visualize a relação entre 2 variáveis (por exemplo idade vs. valor do crédito) usando um gráfico de dispersão.

plt.scatter(df_german['Age'],df_german['Credit amount'])
plt.show()


plt.figure(figsize=(16,8))

#bins = tamanho das barras
plt.scatter(df_german['Age'],df_german['Credit amount'],alpha=0.5)
plt.title("Grafico de dispersao: idade vs Valor do credito")
plt.xlabel('Idade')
plt.ylabel('Valor do Credito')
plt.show()
```

```python
#4. **Gráfico de Boxplot:**
 #  Crie um boxplot para analisar a distribuição de variáveis (por exemplo, distribuição de idade por sexo).
plt.figure(figsize=(16,8))
df_male = df_german[df_german['Sex'] == 'male']['Age']
df_female = df_german[df_german['Sex'] == 'female']['Age']
labels=['Masculino','Feminino']
plt.title("Comparação idade por sexo")
plt.xlabel('Sexo')
plt.ylabel('Idade')
plt.boxplot([df_male,df_female],labels=['Masculino','Feminino'])
plt.show()

```
